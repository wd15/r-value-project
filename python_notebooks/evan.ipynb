````python
import pandas
````

We're trying to read a CSV file that looks like this,

```
# PLS1(9) - XAS
wavelength,7012,7014.5,7017,7019.5,7022,7024.5,7027,...
%Fe3+,1.35457,1.35457,1.35457,1.35457,1.35457,1.35457,1.35457,...
```

It's not oriented correctly so is weird when it's first read in. Let's try reading it in.

````python
pandas.read_csv('regression_model (1).csv')
````

It's obviously the wrong orientation, so let's transpose

````python
pandas.read_csv('regression_model (1).csv').T
````

The last row is totally foobarred, so let's try and ignore that

````python
pandas.read_csv('regression_model (1).csv', skiprows=1).T
````

Better, but the wavelengths are the indices

````python
pandas.read_csv('regression_model (1).csv', header=None, skiprows=1).T
````

Now have a sensible index, but the column headers are now the first row.

````python
pandas.read_csv('regression_model (1).csv', header=None, skiprows=1, index_col=0).T
````

Almost perfect, but the indices start at 0, which might not matter, but we can fix it anyway

````python
pandas.read_csv('regression_model (1).csv', header=None, index_col=0, skiprows=1).T.reset_index(drop=True)
````

Bingo!
